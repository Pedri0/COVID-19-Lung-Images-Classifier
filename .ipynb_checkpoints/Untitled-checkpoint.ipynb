```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage import io
from skimage.segmentation import random_walker

testImages = [io.imread(file,0) for file in glob.glob("TestImages/*.png")]
read = []
for images in range(len(testImages)):
    image = cv2.cvtColor(testImages[images],cv2.COLOR_GRAY2BGR)
    read.append(image)

for images in range(len(read)):
    segmented = slic(read[images], n_segments = 25, sigma=0)
    fig = plt.figure("Superpixels -- %d segments"%(images),figsize=(8,8))
    ax = fig.add_subplot(1, 1, 1)
    ax.imshow(mark_boundaries(read[images], segmented))    
plt.show()
```

```python
#function to show n images
def showImages(images, rows, cols):
    plt.figure(figsize = (16,16))
    for i in range (rows*cols):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title('Lung')
        plt.xticks(())
        plt.yticks(())
```

```python
img = cv2.imread('TestImages/9.png',0)
ret, thresh1 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)      
  
# the window showing output image          
# with the corresponding thresholding          
# techniques applied to the input image     
cv2.imshow('Otsu Threshold', thresh1)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.figure(figsize = (16,16))
plt.imshow(thresh1,cmap='gray')
plt.show()
```

```python
threshold = []
for image1 in (testImages):
    ret, thresh1 = cv2.threshold(image1, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    threshold.append(thresh1)

showImages(threshold,2,5)
```

```python
threshold1 = []
for images in range(len(threshold)):
    image = cv2.cvtColor(threshold[images],cv2.COLOR_GRAY2BGR)
    threshold1.append(image)

for images in range(len(threshold1)):
    segmented = slic(threshold1[images], n_segments = 25, sigma=0,compactness=10)
    fig = plt.figure("Superpixels -- %d segments"%(images),figsize=(8,8))
    ax = fig.add_subplot(1, 1, 1)
    ax.imshow(mark_boundaries(threshold1[images], segmented))    
plt.show()
```
